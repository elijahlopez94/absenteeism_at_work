```python
def cleanColumns(df):
    clean_columns = []
    for column in df.columns:
        clean_column = column.replace(' ', '_').lower()
        clean_columns.append(clean_column)
    df.columns = clean_columns
    return df

cleanColumns(df)
```

```python
def chartValues(df, col):
    feature = df[col]
    value_counts = feature.value_counts().sort_index()
    x = value_counts.index
    y = value_counts.values

    plt.plot(x,y)
    plt.title(f"Count of Individuals vs {col}")
    plt.xlabel(f"{col}")
    plt.ylabel(f"Count")
    return plt.show();

for col in df.columns:
    chartValues(df, col)
```
